## 1. String Handling 
- Change all string to lowercase for uniformity

```python
data = data.applymap(lambda s:s.lower() if type(s) == str else s)
data.head()
```

```python
data.info()
```

## 2. Converting binary variables to 0s and 1s


```python
data['Parking Facility'].unique()
```

```python
data['Security'].unique()
```

```python
data['Status of Possession'].unique()
```

```python
data['Photo Available'].unique()
```

```python
data['Parking Facility'] = data['Parking Facility'].map({'yes':1, 'no':0})
data['Security'] = data['Security'].map({'yes':1, 'no':0})
data['Status of Possession'] = data['Status of Possession'].map({'ready to move':1, 'ready to move in':1, 'under construction':0})
data['Photo Available'] = data['Photo Available'].map({'yes':1, 'no':0})
data.head()
```

```python
data['Brokerage'].unique()
```

```python
data['Transaction Type'].unique()
```

```python
data['Transaction Type'] = data['Transaction Type'].replace('new properity', 'new property')
data['Transaction Type'].unique()
```

```python
data['Brokerage'] = data['Brokerage'].map({'owner':1, 'broker':0})
data['Transaction Type'] = data['Transaction Type'].map({'new property':1, 'resale':0})
```

```python
data['Brokerage'].unique()
```

```python
data['Transaction Type'].unique()
```

## 5. Create derived column 'Actual Price' and 'Extra Cost' from 'Area (sq.ft)' and 'Rs. per sq.ft'

- Calculating 'Actual Price' by multiplying 'Area (sq.ft)' and 'Rs. per sq.ft'
- Finding the difference between 'Price' (from input data)  and 'Actual Price' to find 'Additional Fee'

```python
data['Actual Price'] = data['Area (sq. ft)'] * data['Rs. per sq.ft']
```

```python
data.head()
```

```python
data['Additional Fee'] = (data['Price (Crores)'] * 10000000) - data['Actual Price']
```
