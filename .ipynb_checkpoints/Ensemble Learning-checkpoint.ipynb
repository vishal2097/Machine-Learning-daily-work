```python
from sklearn.datasets import load_iris

import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

from mlxtend.plotting import plot_decision_regions
```

```python
data = load_iris()
```

```python
data.feature_names
```

```python
data.target_names
```

```python
d = {'sepal length (cm)': data.data[50:,1],
     'petal length (cm)': data.data[50:,1],
      'Species':np.array([i for i in data.target if i > 0])
}
```

```python
df =pd.DataFrame(d)
```

```python
df
```

```python
df_train = df.iloc[:60,:].sample(10)
```

```python
df_train
```

```python
df_test = df.iloc[61:,:].sample(10)
```

```python
df_tt = df_test.sample(5, replace =True)
x_test = df.iloc[:,:-1].values
y_test = df.iloc[:,-1].values
```

```python
def evaluate(model, x, y):
    model.fit(x,y)
    plot_tree(model)
    
    plt.show()
    plot_decision_regions(x,y,model,legend =2)
    plt.show()
    
    y_pred = model.predict(x_test)
    
    print("y_test:", y_test)
    print("y_pred :",y_pred)
    
    print("accuracy score",accuracy_score(y_test,y_pred))
    
    return model
```

```python
#first decision tree
df_t = df_train.sample(8, replace =True)
x_train = df.iloc[:,:-1].values
y_train = df.iloc[:,-1].values

dt = DecisionTreeClassifier()
bag1 = evaluate(dt, x_train, y_train)
```
