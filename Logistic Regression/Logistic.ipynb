# Question:-prediction of children gender using Age and cartoon_show Atrribute

```python
#Import the required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('ML_dataset.xlsx')
```

```python
data
```

```python
#Handling missing values
data.info()
```

```python
data = data.applymap(lambda s:s.lower() if type(s) == str else s)
data.head()
```

```python
data.describe()
```

```python
data['cartoon_show'].unique()
```

```python
data['cartoon_show'] = data['cartoon_show'].map({'pepapig':0, 'shinchan':1,'legen kora':2, 'doremon':3 })

#mapping of cartoon_show datapoints.
```

```python
print(data.head())
print(data.shape)
```

```python
# Set the predictor(X) and Response(Y) Variables

X = data[['Age','Standard','cartoon_show']]
Y = data['Gender']
```

```python
# spliting into training and Test Set
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test = train_test_split(X,Y, test_size=0.25, stratify = Y,random_state=25)
```

# Logistic regression

```python
# call logistic regression Model
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
logit.fit(X_train,Y_train)
```

```python
# compute the Predictions or Y hat
Y_pred = logit.predict(X_test)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
print("confusion matrix for logistic regression\n",cm)
```

```python
#Accuracy Score
from sklearn.metrics import accuracy_score
lr_accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy of model using Logistic Regression is ",lr_accuracy)
```
