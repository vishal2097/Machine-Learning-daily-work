# In this file using following columns build a model to predict if person would survive or not,

Pclass

Sex

Age

Fare


```python
#import libaroes
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()
```

```python
#create on dataframe of necessary features
inputs = df[['Pclass','Sex','Age','Fare']]
```

```python
inputs
```

```python
target = df['Survived']
target
```

```python
inputs.info()
```

```python
#Encode the sex attribute as m =1 and F =0
from sklearn.preprocessing import LabelEncoder
Sex_n = LabelEncoder()
inputs['Sex_'] = Sex_n.fit_transform(inputs['Sex'])
```

```python
inputs
```

```python
#drop unencoded feautres
inputs_n = inputs.drop(['Sex'], axis ='columns')
```

```python
inputs_n
```

```python
#ti fill nan vallues of age with age mean
inputs_n.Age = inputs_n.Age.fillna(inputs_n.Age.mean())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(inputs_n,target,test_size=0.2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

# Conclusion

The above model of decision tree got 81 % accuracy
